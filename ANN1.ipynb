```
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/cancerdata-cleanedwell.csv")
```

```
df.head(10)
```

```
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
print(x)
print(y)
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

```

```
from sklearn.preprocessing import StandardScaler
scaledata=StandardScaler()
x_train=scaledata.fit_transform(x_train)
x_test=scaledata.fit_transform(x_test)
```

```
from keras.models import Sequential
model=Sequential()
#building layers
from keras.layers import Dense   #dense layer only used in ANN
from keras.layers import Dropout
#to increase the accuracy we use dropout
#dropout can be added before input layer and hidden layer
model.add(Dropout(0.3,input_dim=31))

#creating input layer
Dense(units=31,kernel_initializer='he_uniform',input_dim=31)
```

```
model.add(Dropout(0.3))
#creating hidden layer
model.add(Dense(units=31*1.5,kernel_initializer='he_uniform',activation='relu'))
```

```
#creating o/p layer
model.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```
from tensorflow import keras
opt=keras.optimizers.Adam(learning_rate=0.01)
#compiler
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy']) #to increase the accuracy
#model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
print(model)
```

```
model.summary
```

```
#validation
model_hist=model.fit(x_train,y_train,validation_split=0.25,batch_size=30,epochs=100)

#parameter tuning is use to increase accuracy
#model_hist=model.fit(x_train,y_train,validation_split=0.25,batch_size=30,epochs=100)
```

```
model_hist.history['accuracy']
print(model_hist.history['accuracy'])
```

```
#Test
ycap=model.predict(x_test)
ycap=(ycap>0.5)
score=accuracy_score(y_test,ycap)
print(score)
```
